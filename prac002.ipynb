```python
import pandas as pd
import numpy as np

data_dirs = ['./data/murphy/learn','./data/murphy/test']

original_data = pd.DataFrame()

original_data.shape
```

```python
import os 

for dd in data_dirs:
    filenames = os.listdir(dd)
    filenames.sort()
    print("{}안에는 {}개의 파일이 존재".format(dd, len(filenames)))
    print(filenames[:10])
```

```python
targetfilepath = './data/prac2.data'
instance_id = 0
for dd in data_dirs:
    filenames = os.listdir(dd)
    filenames.sort()
    print("{}안에는 {}개의 파일 존재".format(dd, len(filenames)))
    for fn in filenames:
        full_path = dd + '/' + fn
        with open(full_path, 'r')as rf:#읽기 전용으로 full_path 읽어서 rf파일
            while True:
                content = rf.readline()#rf 파일 한줄을 읽어서 content에 넣는다.
                if not content : break #content 가 없으면 멈춘다. 
                while '  'in content:#content안에 두칸의 공백이 존재하면
                    content = content.replace('  ', ' ')#두칸짜리 공백을 한칸으로 교체한다.
                content = '{}'.format(instance_id) + content# 끝에 줄바꿈, 문자 존재 숫자를 문자열로 바꿨다(형바꿈)
                with open(targetfilepath, 'a') as wf:#a모드로 wf
                    wf.write(content)
        instance_id += 1
```

```python
original_data = pd.read_csv(targetfilepath, sep=' ', dtype='unicode', \
                           names=['id', 'class', 'objnum', 'type', 'xx1',\
                                 'yy1', 'xx2', 'yy2', 'size', 'diag'])
print(original_data.shape)
original_data.head()
```

```python
pd.crosstab(original_data['class'], original_data.objnum, margins=True)
```

```python
ids = original_data['id'].unique()
print(ids)
```

```python
tmp_feat_names = {'id': [], 'class': [], \
                 '0_xx1': [], '0_yy1': [], '0_xx2': [], '0_yy2': [], '0_size': [],'0_diag': [], \
                 '1_xx1': [], '1_yy1': [], '1_xx2': [], '1_yy2': [], '1_size': [],'1_diag': [], \
                 '2_xx1': [], '2_yy1': [], '2_xx2': [], '2_yy2': [], '2_size': [],'2_diag': []}
filtered_data = pd.DataFrame(tmp_feat_names)
filtered_data.head()

ids = [int(x) for x in ids]
print(ids)
```

```python
for i in ids:
    if i % 100 == 0: print(i)
    tmp_df = original_data[original_data['id'] =='{}'.format(i)]
    tmp_instance = tmp_feat_names.copy()
    tmp_instance['id'] = tmp_df['id'].values[0]
    tmp_instance['class'] = tmp_df['class'].values[0]
        
    from_feats = ['diag', 'size', 'xx1', 'yy1', 'xx2', 'yy2']
    for obj_idx in range(0, 3):
        to_feats = ['{}_{}'.format(obj_idx, x)for x in from_feats]
        tmp_series = tmp_df[tmp_df.objnum =='{}'.format(obj_idx)]
        for j in range(len(from_feats)):
            tmp_instance[to_feats[j]] = tmp_series[from_feats[j]].values[0]
    filtered_data = filtered_data.append(tmp_instance, ignore_index=True)
        
```

```python
filtered_data.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.naive_bayes import MultinomialNB

features = ['0_diag', '1_diag', '2_diag', '0_size', '1_size', '2_size', \
           '0_xx1', '1_xx1', '2_xx1', '0_xx2', '1_xx2', '2_xx2', \
           '0_yy1', '1_yy1', '2_yy1', '0_yy2', '1_yy2', '2_yy2']
kf = KFold(n_splits=5, shuffle=True)
print(type(kf))
fold_idx = 1
for train_index, test_index in kf.split(filtered_data):
    print('[Fold{}] train size={}, test size={}'.format(fold_idx, len(train_index), len(test_index)))
    train_data, test_data = filtered_data.iloc[train_index], filtered_data.iloc[test_index]
    
    train_y = train_data['class']
    train_x = train_data[features]
    
    train_y = test_data['class']
    train_x = test_data[features]
    
    model = MultinomialNB()
    model.fit(train_x, train_y)
    
    mean_accuracy =model.score(test_x, test_y)
    print('mean_accuracy = {}'.format(mean_accuracy))
    
    fold_idx += 1
    
```
